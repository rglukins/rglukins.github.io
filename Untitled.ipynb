```python
import pandas as pd
import json
import csv
```

```python
with open('city_data.json') as file:
    data = json.load(file)
```

```python
data
```

```python
city_data = data
```

```python
humidity = [city['main']['humidity'] for city in city_data]
ws = [city['wind']['speed'] for city in city_data]
country = [city['sys']['country'] for city in city_data]
longitude = [city['coord']['lon'] for city in city_data]
latitude = [city['coord']['lat'] for city in city_data]
temperature = [city['main']['temp'] for city in city_data]
cloudiness = [city['clouds']['all'] for city in city_data]
city = [city['name'] for city in city_data]

final_cities_df = pd.DataFrame({
    'Name': city,
    'Country': country,
    'Latitude': latitude,
    'Longitude': longitude,
    'Temperature': temperature,
    'Humidity': humidity,
    "% Cloudiness": cloudiness,
    'Wind Speed': ws
})

final_cities_df.head()
```

```python
final_cities_df = final_cities_df[["Name", 'Country', 'Longitude', 'Latitude', 'Temperature', 'Humidity', 'Wind Speed', '% Cloudiness']]
```

```python
final_cities_df.to_html('equator_data.html', index=False, header=True)
```
